### Visualizations

```python
#Visualization- Map of the airports included in the dataset based on their coordinates

gps_lat = []
gps_long = []
latt = []
long = []

# Properly formats the Gps coordinated for later use
for key, value in AirportCode.Airport_codes.items():
    coor = value[1].split(", ")
    gps_lat.append(coor[0])
    gps_long.append(coor[1])

# Defines a function that converts gps coordinate
# to either lattitude or longitude coordinate
def gpsToLatLong(old):
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    newcoor = old.replace(u'°',' ')
    newcoor = newcoor.replace('′',' ')
    newcoor = newcoor.replace('″',' ')
    newcoor = newcoor.replace('`',' ')
    newcoor = newcoor.split()
    new_dir = newcoor.pop()
    newcoor.extend([0,0,0])
    return (int(newcoor[0])+int(newcoor[1])/60.0+int(newcoor[2])/3600.0) * direction[new_dir]

# Converts gps coordinates and stores them into latt and long arrays
for i in range(0, len(gps_lat)):
    latt.append(gpsToLatLong(gps_lat[i]))
    long.append(gpsToLatLong(gps_long[i]))
    
# Converts the latt and long coordinates into a projection
# the x and y coordinate plane
xcoor = []
ycoor = []
for i in range(0,  len(latt)):
    xcoor.append((100/360.0) * (180 + long[i]))
    ycoor.append((100/180.0) * (90 - latt[i]));

    
# Plots the coordinates of airports using 
# the x and y coordinate planes
plt.scatter(xcoor,ycoor)

plt.title('Map of all airports')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

### Decision Tree Classifier To Predict Number of Connecting Flights

```python
#Decision Tree Classifier To Predict Number of Connecting Flights

df_fare = pd.read_csv('UPDATEDitineries-airportCodes3.csv', low_memory=False)
X = df_fare[['isBasicEconomy','isRefundable','minutes','seatsRemaining','baseFare','destinationAirportDistance','startingAirportDistance', 'totalTravelDistance', "totalFare"]]
Y = df_fare["connections"] 
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.4, random_state=42)
model = tree.DecisionTreeClassifier(max_depth = 5)
model.fit(x_train, y_train)
predictions_test = model.predict(x_test)
predictions_train = model.predict(x_train)
accuracy_train = accuracy_score(y_train, predictions_train)
accuarcy_test = accuracy_score(y_test, predictions_test)
print("Decision Tree - Connections")
print("Train")
print(accuracy_train)
print("test")
print(accuarcy_test)
```

In this decisionTreeClassifier we were able to predict the number of connections with a training accuracy train of 92% and an accuracy test score of 91.8%. This decisionTreeClassifier was able to become much more accurate as we turned our starting and ending airports string codes into number representation whereas they became floats as the distance between the airport and Atlanta. We believe this model was so successful as it was able to find the strong correlation between the minutes and distance of total travel time and the actual distance between the airports. We also think it found the correlation between the price and the seat remaining on flights with a large amount of connection. Whereas there are only 4 options for total connections - 0,1,2,3 - if the price was lower than other flights between the same airports and the seats remaining were larger than it could more easily find outliers. We think the decision tree was the best implementation in this example as with the limited amount of possible outputs it's good to have a model that is able to easily split data that has a out of proportion 

### K-Nearest Neighbors Model To Predict Destination Airport

### Hierarchical Clustering Analysis

```python
#Hierarchical Clustering Analysis

df_fare = pd.read_csv('UPDATEDitineries-airportCodes3.csv', low_memory=False)
df_fare = df_fare.select_dtypes(include='number')
bin_size = 75
df_fare_binned = df_fare.median().to_frame().transpose()  # Compute median of each column
df_fare_binned = df_fare.groupby(pd.cut(df_fare.index, bins=range(0, len(df_fare.index), bin_size))).median()

#HCA with minutes and totalFare

df_fare_3 = df_fare_binned[['minutes','totalFare']]
df_fare_3_nm = StandardScaler().fit_transform(df_fare_3)

plt.scatter(df_fare_3_nm[:,0], df_fare_3_nm[:, 1])

# create a dendrogram
clust = linkage(df_fare_binned)
dendrogram(clust)
plt.show()

from scipy.cluster.hierarchy import fcluster

# use fcluster to visualize the clusters better
fclust = fcluster(clust, 3, criterion = "maxclust")
plt.scatter(df_fare_3_nm[:, 0], df_fare_3_nm[:, 1], c=fclust)
```

```python
#HCA with totalTravelDistance and connections

df_fare_5 = df_fare_binned[['minutes','connections']]
df_fare_5_nm = StandardScaler().fit_transform(df_fare_5)

plt.scatter(df_fare_5_nm[:,0], df_fare_5_nm[:, 1])

# create a dendrogram
clust = linkage(df_fare_binned)
dendrogram(clust)
plt.show()

from scipy.cluster.hierarchy import fcluster

# use fcluster to visualize the clusters better
fclust = fcluster(clust, 3, criterion = "maxclust")
plt.scatter(df_fare_5_nm[:, 0], df_fare_5_nm[:, 1], c=fclust)
```

We created unsupervised learning models using Hierarchical Clustering Analysis to see if distinct clusters would formed when combinations of two features were passed into the model. The two models that ended up being the most relevant used minutes and total fare as features, and minutes and connections as features. In the first model, three clusters appear. The clusters suggests that past a certain flight time, flights group up into a small range of prices, and under a certain flight time, flights group into a small range of prices. In between the extreme groups, it appears that flight time and price have a somewhat linear relationship. In the second model, the clusters gather into distinct lines because the number of potential connecting flights of the data points has a small range. However, one can see that the distribution of the clusters in the second model matches the width of the clusters in the first model, suggesting that the three "groups" that appear based on flight time and price also correlate to the number of connections per flight.
